```python
Q1 = df['radius_mean'].quantile(0.25)
Q3 = df['radius_mean'].quantile(0.75)
IQR = Q3 - Q1
outliers = df[(df['radius_mean'] < Q1 - 1.5 * IQR) | (df['radius_mean'] > Q3 + 1.5 * IQR)]
print("Outliers in radius_mean:\n", outliers[['radius_mean']])
```

```python
Q5 = df['texture_worst'].quantile(0.25)
Q7 = df['texture_worst'].quantile(0.75)
IQR = Q7 - Q5
outliers = df[(df['texture_worst'] < Q5 - 1.5 * IQR) | (df['texture_worst'] > Q7 + 1.5 * IQR)]
print("Outliers in texture_worst:\n", outliers[['texture_worst']])
```

```python
Q9=df['perimeter_mean'].quantile(0.25)
Q11=df['perimeter_mean'].quantile(0.75)
IQR=Q11-Q9
outliers=df[(df['perimeter_mean']<Q9-1.5*IQR)|(df['perimeter_mean']>Q11+1.5*IQR)]
print("outliers in perimeter_mean:\n",outliers[['perimeter_mean']])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 5))
sns.boxplot(x='diagnosis', y='radius_mean', data=df)
plt.title('Box Plot of Radius Mean by Diagnosis')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 5))
sns.boxplot(y='perimeter_mean', x='diagnosis', data=df)
plt.title('Box Plot of perimeter Mean by Diagnosis')
plt.show()
```

```python
sorted_vals = np.sort(df['perimeter_mean'])
cdf = np.arange(1, len(sorted_vals)+1) / len(sorted_vals)

plt.plot(sorted_vals, cdf, marker='.')
plt.title("CDF of Perimeter Mean")
plt.xlabel("Perimeter_mean")
plt.ylabel("CDF")
plt.grid(True)
plt.show()
```

```python
numeric_df = df.select_dtypes(include=[np.number])
corr = numeric_df.corr()
print("Correlation with radius_mean:\n", corr['radius_mean'].sort_values(ascending=False))

```

```python
df['diagnosis_code'] = df['diagnosis'].map({'M': 1, 'B': 0})
numeric_df = df.select_dtypes(include=[np.number])
corr = numeric_df.corr()
print("Correlation with diagnosis_code:\n", corr['diagnosis_code'].sort_values(ascending=False))
```

```python
from scipy.stats import pearsonr
pearson_corr, _ = pearsonr(df['radius_mean'], df['diagnosis_code'])
print("Pearson correlation (radius_mean vs diagnosis):", pearson_corr)
```

```python
from scipy.stats import ttest_ind
```

```python
benign_r = df.loc[df['diagnosis'] == 'B', 'radius_mean'].dropna()
malignant_r = df.loc[df['diagnosis'] == 'M', 'radius_mean'].dropna()
t_stat, p_value = ttest_ind(benign_r, malignant_r, equal_var=False)
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4e}")
if p_value < 0.05:
    print("Result: Statistically significant difference in radius_mean between Benign and Malignant.")
else:
    print("Result: No statistically significant difference.")

```
