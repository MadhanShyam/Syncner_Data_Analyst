```python
df = df.dropna(subset=['Survived'])
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Fare'] = df['Fare'].fillna(df['Fare'].median())
```

```python
numerical_cols = ['Age', 'Fare', 'SibSp', 'Parch']
categorical_cols = ['Sex', 'Pclass', 'Embarked']
```

```python
from scipy.stats import ttest_ind
from statsmodels.stats.weightstats import ztest

print("T-Test (Numerical vs Survived)\n")
for col in numerical_cols:
    group1 = df[df['Survived'] == 1][col]
    group0 = df[df['Survived'] == 0][col]
   
    t_stat, p_val_t = ttest_ind(group1, group0)
    sig_t = "Significant" if p_val_t < 0.05 else "Not significant"
    
    z_stat, p_val_z = ztest(group1, group0)
    sig_z = "Significant" if p_val_z < 0.05 else "Not significant"
    
    print(f"{col} â†’ T-test p = {p_val_t:.4f} ({sig_t}), Z-test p = {p_val_z:.4f} ({sig_z})")
```

```python
from scipy.stats import chi2_contingency

print("\n Chi-Square Test (Categorical vs Survived)\n")
for col in categorical_cols:
    contingency = pd.crosstab(df[col], df['Survived'])
    chi2, p_val_chi, dof, expected = chi2_contingency(contingency)
    sig = " Significant" if p_val_chi < 0.05 else " Not significant"
    print(f"{col} â†’ Chi-square p = {p_val_chi:.4f} ({sig})")
```
